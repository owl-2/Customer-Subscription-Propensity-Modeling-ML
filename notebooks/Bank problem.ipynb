```python
param_dist = {'class_weight':['balanced'], 
        'criterion':['entropy','gini','log_loss'], ## added log_loss
        'max_depth':[None,5,10,15,20,30,50,70],
        'min_samples_split':[20,25,30,35,40,45,50,55,70],
        'min_samples_leaf':[5,10,15,20,25,30,35,40,], # Till here, its same as DT
              
        "n_estimators":[100,200,300,500], # Number of trees to build
        "max_features": [6,7,9,12,15,17],
          "max_samples": [None, 0.7, 0.8]
         }
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        # np.flatnonzero(): Return the indices of the non-zero elements
        
        for candidate in candidates:
            print(f"Model with rank: {i}")
            print(f"Mean validation score: {round(results['mean_test_score'][candidate], 5)} \
                  std: {round(results['std_test_score'][candidate], 5)}")
            print(f"Parameters: {results['params'][candidate]}")
            print("")
```

```python
random_search.best_estimator_
```

```python
report(random_search.cv_results_,3)
```

```python
def detect_outliers_iqr_df(bank_train, columns):
    outliers = {}
    for col in columns:
        q1, q3 = np.percentile(bank_train[col], [25, 75])
        iqr = q3 - q1
        lower_bound = q1 - (1.5 * iqr)
        upper_bound = q3 + (1.5 * iqr)
        outliers[col] = bank_train[(bank_train[col] < lower_bound) | (bank_train[col] > upper_bound)][col]
    return outliers
```

```python
# Detect outliers in specified columns
columns_to_check = ['balance']  # Specify the columns to check for outliers
outliers = detect_outliers_iqr_df(bank_train, columns_to_check)
print("Outliers:", outliers)
```
